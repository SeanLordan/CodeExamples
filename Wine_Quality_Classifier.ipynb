```python
f1 = 0

f2 = 1

f3 = 2

 

desc = dict(

    classes = [1, 0],

    colors  = ["orange", "blue"],

    labels  = ['Low Quality ("1")', 'High Quality ("0")'],

    symbols = ["x", "circle"]

)

 

data = [

    Scatter3d(

        x = X[y == desc["classes"][i], f1],

        y = X[y == desc["classes"][i], f2],

        z = X[y == desc["classes"][i], f3],

        name = desc["labels"][i],

        mode = "markers",

        marker = dict(

            size = 2.5,

            symbol = desc["symbols"][i],

            color  = desc["colors"][i]

        )

    ) for i in range(len(desc["labels"]))

]

 

layout = Layout(

    scene=Scene(

        xaxis=XAxis(title=header[f1], titlefont=dict(size=11)),

        yaxis=YAxis(title=header[f2], titlefont=dict(size=11)),

        zaxis=ZAxis(title=header[f3], titlefont=dict(size=11))

    ),

    margin=Margin(l=80, r=80, b=0, t=0, pad=0, autoexpand=True),

    height= 600,

)

 

fig = dict(data = data, layout = layout)

 

py.iplot(fig)
```

```python
# now lets have a scatter plot for each combination!

from plotly import tools

 

# so as not to be too crowded we are creating a grid of subplots 5 x 5

fig = tools.make_subplots(rows = 11, cols = 11, shared_xaxes = True, shared_yaxes = True)

 

for row in range(0, 10):  # here we define the 4 row50):  # and the 4 cols

        # low quality scatter (class 1) represented with an orange marker

        trace1 = Scatter(

            # based on our analysis and histogram we might assign 7

            # as the cutoff between good and bad Wined (remember Y is our label)

            x = X[y < 8, col],  # Y refers to the Y Label created right at the beginning (X are the features)

            y = X[y < 8, row],

            mode = 'markers',

            name = 'Poor Quality ("1")',

            marker = dict(

                color = 'orange',

                symbol = 'x',

                opacity = .5

            )

        )

        # High quality scatter (class 1) represented with an orange marker

        trace2 = Scatter(

            # based on our analysis and histogram we might assign 7

            # as the cutoff between good and bad Wined (remember Y is our label)

            x = X[y >= 8, col],  # Y refers to the Y Label created right at the beginning (X are the features)

            y = X[y >= 8, row],

            mode = 'markers',

            name = 'High Quality ("0")',

            marker = dict(

                color = 'blue',

                symbol = 'circle',

                opacity = .5

            )

        )

        posX = row+1

        posY = col+1

        fig.append_trace(trace1, posX, posY)

        fig.append_trace(trace2, posX, posY)

        fig['layout']['xaxis'+str(posX)].update(title=header[row])

        fig['layout']['yaxis'+str(posY)].update(title=header[col])

       

fig['layout'].update(

    showlegend = False,

    height = 900

)

py.iplot(fig)
```
